## Gráficos de Línea

```python
import matplotlib.pyplot as plt

meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
ventas = [120, 150, 135, 170, 200, 250, 220, 240, 280, 300, 280, 320]

plt.plot(meses, # x: Variable independiente
         ventas, # y: Variable dependiente
         marker='^', # Ciruclo: 'o', Cuadrado: 's', Diamante: 'D', Triangulo_Arriba: '^', y más....
         linestyle=':', # Solida: '-', Rayas:'--', Punteada:':', Sin_Linea: '', y más...
         color='#FF5733')  # Rojo: 'red', Hexadecimal:'FF5733'
plt.title('Ventas Mensuales')
plt.xlabel('Meses')
plt.ylabel('Ventas (en miles de dólares)')

plt.show()
```

## Gráficos de dispersión

```python
import matplotlib.pyplot as plt

tiempo_estudio = [1.5, 2.0, 2.5, 1.0, 3.0, 2.5, 4.0, 3.5, 2.0, 1.5]
calificaciones = [60, 65, 75, 55, 80, 70, 85, 75, 60, 50]

plt.scatter(tiempo_estudio, calificaciones, color='blue')
plt.title('Relación entre Tiempo de Estudio y Calificaciones')
plt.xlabel('Tiempo de Estudio diario (horas)')
plt.ylabel('Calificaciones en el examen (porcentaje)')

plt.show()
```

# Barras de Error

### Barras de error simétricas utilizando el argumento yerr:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
error = [0.5, 0.3, 0.7, 0.2, 0.4]

plt.errorbar(x, y,
             yerr=error, # Rango de error vertical
             fmt='p') # Fotmato de linea y marcado, ej: 'o-.', '-'
plt.show()
```

### Barras de error asimétricas utilizando los argumentos yerr y xerr:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
x_error = [0.1, 0.2, 0.3, 0.1, 0.2]
y_error = [0.5, 0.3, 0.7, 0.2, 0.4]

plt.errorbar(x, y, xerr=x_error, yerr=y_error, fmt='o')
plt.show()
```

### Ejemplo

```python
import matplotlib.pyplot as plt

condiciones_luz = ['Baja', 'Media', 'Alta']
altura_promedio = [10, 15, 20]
errores_altura = [2, 3, 4]

plt.errorbar(condiciones_luz, altura_promedio, yerr=errores_altura, fmt='o',
             capsize=5) #  Longitud de líneas horizontales en los extremos
plt.title('Altura de Plantas en Diferentes Condiciones de Luz')
plt.xlabel('Condiciones de Luz')
plt.ylabel('Altura de las Plantas (cm)')

plt.show()
```

## Con etiquetas

```python
import matplotlib.pyplot as plt

edades = [22, 25, 30, 33, 35, 38, 40, 40, 42, 45, 48, 50, 52, 55, 60, 62, 65, 70, 72, 75]

plt.hist(edades, bins=5,
         edgecolor='black')         # Color del borde de las barras
plt.title('Distribución de Edades') # Titulo del Grafico
plt.xlabel('Edades')                # Descripcion del eje horizontal
plt.ylabel('Frecuencia')            # Descripción del eje vertical

plt.show()
```

## Comparamos graficos y agregamos leyendas

```python
import matplotlib.pyplot as plt

# Lista original curso A
edades_A = [22, 25, 30, 33, 35, 38, 40, 40, 42, 45, 48, 50, 52, 55, 60, 62, 65, 70, 72, 75]

# Crear una nueva lista de edades con valores ligeramente inferiores
edades_B = [22, 23, 25, 31, 33, 37, 39, 40, 42, 42, 48, 50, 52, 55, 55, 55, 57, 58, 58, 75]

# Graficar ambos histogramas en el mismo gráfico
plt.hist(edades_A, bins=5, edgecolor='black',
         alpha=0.7,         # Opacidad o transparencia del grafico, va de 0 a 1
         label='Edades A')  # Agregar leyenda sobre el gráfico, pero sin mostrarlo

plt.hist(edades_B, bins=5, edgecolor='black', alpha=0.7, label='Edades B')

plt.title('Comparación de Distribuciones de Edades')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
# Mostramos la leyenda
plt.legend()
# Personalizamos la leyenda
plt.legend(loc='upper right', # Ubicación ej: lower left
           fontsize='small')  # Tamaño ej: medium, 12, 20
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Trazar las líneas y agregar etiquetas
plt.plot(x, y1, label='Línea 1')
plt.plot(x, y2, label='Línea 2')

# Agregar leyenda al gráfico
plt.legend()

# Personalizar la leyenda
plt.legend(loc='upper left', fontsize='small')

# Mostrar el gráfico
plt.show()
```
